## Data engineering
The functions was created and used to clean the data 

```python
# Function to rearrange the columns
def re_arange_column(column,df):
                    df.insert(len(df.columns)-1, column, df.pop(column))
                    return df
```

```python
# Function to remove null and replace it with the empty space
def remove_nall(column,df):
                    df[[column]]=df[[column]].fillna('')
                    return df
```

```python
# Function to delete columns
def remove_col(column,df):
                    del df[column]
                    return df
```

```python
# This function help to replace values in a column
def replacing_values(Old_value,New_value,column):
                    column.replace(['Old_value'],'New_value')
                    return column
```

```python
# Function to fill up values in column useing other column information
def opposit_value_col(column1, column2,df):
                    if df[column1] == 'FALSE':
                                        return df[column2]== 'FALSE'
                    else:

                                        return df[column2] == 'TRUE'
```
